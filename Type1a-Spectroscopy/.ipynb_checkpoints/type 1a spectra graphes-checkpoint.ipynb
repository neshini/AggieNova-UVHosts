```python
#L=F*D^2
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# SN2011by, NGC3972
#distance in cm
d_a=6.05893E+25
file_a = 'SN2011by_110509_uvopt_obswave_obsflux.dat'
f_a= pd.read_csv(file_a, sep='\s+')

#used to index row,column; part of pandas
flux_a = f_a.iloc[:, 1]
wave_a = f_a.iloc[:,0]

lum_a = flux_a*(d_a**2)
wave_a
```

```python
# SN2011iv, NGC1404
d_b=6.02554E+25
file_b = 'SN2011iv_111211_full.dat'
f_b= pd.read_csv(file_b, sep='\s+')

flux_b = f_b.iloc[:, 1]
wave_b = f_b.iloc[:,0]

lum_b = flux_b*(d_b**2)
wave_b
```

```python
# SN2011fe, M101
d_c=6.1291E+25
file_c = 'SN2011fe_110907_uvopt_obswave_obsflux.dat'
f_c= pd.read_csv(file_c, sep='\s+', nrows= 2553)

flux_c = f_c.iloc[:, 1]
wave_c = f_c.iloc[:,0]

lum_c = flux_c*(d_c**2)
wave_c
```

```python
# SN2016ccj, NGC4125
d_d=2.25278E+26
file_d = 'SN2016ccj_160514_uvopt_obswave_obsflux.dat'
f_d= pd.read_csv(file_d, sep='\s+', nrows=450)

flux_d = f_d.iloc[:, 1]
wave_d = f_d.iloc[:,0]

lum_d = flux_d*(d_d**2)
wave_d
```

```python
# SN2017erp, NGC5861
d_e=8.62573E+25
file_e = 'SN2017erp_170702_uvopt_obswave_obsflux.dat'
f_e= pd.read_csv(file_e, sep='\s+')

flux_e = f_e.iloc[:, 1]
wave_e = f_e.iloc[:,0]

lum_e = flux_e*(d_e**2)
wave_e
```

```python
plt.figure()

plt.xlabel('flux density')
plt.ylabel('luminosity')

plt.plot(flux_a, lum_a, label='SN2011by')
plt.plot(flux_b, lum_b, label='SN2011iv')
plt.plot(flux_c, lum_c, label='SN2011fe')
plt.plot(flux_d, lum_d, label='SN2016ccj')
plt.plot(flux_e, lum_e, label='SN2017erp')

plt.legend()
plt.show()
plt.close()
```

```python
plt.figure()
plt.xlabel('Flux Density (erg/s/cm^2)')
plt.ylabel('Luminosity (erg/s/Angstrom)')
plt.plot(flux_a, lum_a, label='SN2011by', color='r')
plt.plot(flux_b, lum_b, label='SN2011iv', color='y')
plt.plot(flux_d, lum_d, label='SN2016ccj', color='g')
plt.plot(flux_e, lum_e, label='SN2017erp', color='b')
plt.title("Flux Density vs Luminosity")
plt.legend()
plt.savefig('fluxvlum(4)')
plt.show()
plt.close()

plt.figure()
plt.xlabel('Flux Density (erg/s/cm^2)')
plt.ylabel('Luminosity (erg/s/Angstrom)')
plt.plot(flux_a, lum_a, label='SN2011by', color='r')
plt.plot(flux_b, lum_b, label='SN2011iv', color='y')
plt.plot(flux_c, lum_c, label='SN2011fe', color='m')
plt.plot(flux_d, lum_d, label='SN2016ccj', color='g')
plt.plot(flux_e, lum_e, label='SN2017erp', color='b')
plt.title("Flux Density vs Luminosity")
plt.legend()
plt.savefig('fluxvlum(5)')
plt.tight_layout()
plt.show()
plt.close()
```

```python
#wave x, flux y
plt.figure()
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux Density (erg/s/cm^2)')
plt.plot(wave_a, flux_a, label='SN2011by', color='r')
plt.plot(wave_b, flux_b, label='SN2011iv', color='y')
plt.plot(wave_d, flux_d, label='SN2016ccj', color='g')
plt.plot(wave_e, flux_e, label='SN2017erp', color='b')
plt.title(" Wavelength vs Flux Density")
plt.legend()
plt.savefig('wavevflux(4)')
plt.show()
plt.close()


plt.figure()
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux Density (erg/s/cm^2)')
plt.plot(wave_a, flux_a, label='SN2011by', color='r')
plt.plot(wave_b, flux_b, label='SN2011iv', color='y')
plt.plot(wave_c, flux_c, label='SN2011fe', color='m')
plt.plot(wave_d, flux_d, label='SN2016ccj', color='g')
plt.plot(wave_e, flux_e, label='SN2017erp', color='b')
plt.title(" Wavelength vs Flux Density")
plt.legend()
plt.savefig('wavevflux(5)')
plt.show()
plt.close()
```

```python
#wave x, lum y
plt.figure()
plt.xlabel('Wavelength (A)')
plt.ylabel('Luminosity (erg/s/Angstrom)')
plt.plot(wave_a, lum_a, label='SN2011by', color='r')
plt.plot(wave_b, lum_b, label='SN2011iv', color='y')
plt.plot(wave_d, lum_d, label='SN2016ccj', color='g')
plt.plot(wave_e, lum_e, label='SN2017erp', color='b')
plt.title(" Wavelength vs Luminosity")
plt.legend()
plt.savefig('wavevlum(4)')
plt.show()
plt.close()


plt.figure()
plt.xlabel('Wavelength (A)')
plt.ylabel('Luminosity (erg/s/Angstrom)')
plt.plot(wave_a, lum_a, label='SN2011by', color='r')
plt.plot(wave_b, lum_b, label='SN2011iv', color='y')
plt.plot(wave_c, lum_c, label='SN2011fe', color='m')
plt.plot(wave_d, lum_d, label='SN2016ccj', color='g')
plt.plot(wave_e, lum_e, label='SN2017erp', color='b')
plt.title(" Wavelength vs Luminosity")
plt.legend()
plt.savefig('wavevlum(5)')
plt.show()
plt.close()
```
