```python
#for prez
from PIL import Image
import glob

rows = 10
cols = 9
image_folder = "prop_col_plots"
output_path = "slide_background.png"

image_paths = sorted(glob.glob(f"{image_folder}/*.png"))[:90]  # grab first 36 if you have more
images = [Image.open(p).convert('RGB') for p in image_paths]

w, h = images[0].size
grid_img = Image.new('RGB', (cols*w, rows*h), color='white')

for i, img in enumerate(images):
    row = i // cols
    col = i % cols
    grid_img.paste(img, (col*w, row*h))

grid_img.save(output_path)
print(f"final plot wall: {output_path}")
```
